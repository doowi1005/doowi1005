```python
# max row, columns 셋팅
import pandas as pd
pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 500)
import numpy as np
from pandas import DataFrame, Series
import os
import datetime
import re

######################어도비 데이터 그대로 불러오기####################


sheet=['Country','Referring Domain','Marketing Channel','URL','Article','Download']

semi=Series()
for i in range(0,len(sheet)):
    semi[sheet[i]]= pd.read_excel('C:/Users/cnxk/Desktop/dashboard_prac/Semiconductor_Global_rawdata_jul.xlsx',

                              sheet_name = sheet[i], 

                              header = 1 )   
    
##dimension열 같은 필요없는 컬럼삭제
    col_not_dimension = []
    for j in semi[sheet[i]].columns:
        if bool(re.match('^Dimension',j)) == False:
            col_not_dimension.append(j)
    semi[sheet[i]]=semi[sheet[i]][col_not_dimension]
    
##자이제 Item컬럼의 빈행, 토탈행, low-traffic 행을 지워야한다.    
    semi[sheet[i]]=semi[sheet[i]].dropna(axis=1,how='all')#첫째열은 전체 빈열->삭제해주기
    semi[sheet[i]]=semi[sheet[i]].dropna(how='any')#NA값들어간 행 지우기
    semi[sheet[i]]=semi[sheet[i]].loc[ (semi[sheet[i]]['Item.1']!='Total'),]
    semi[sheet[i]]=semi[sheet[i]].head(20)

#'Referring Domain'->item.1=total

semi['Country']

writer = pd.ExcelWriter('C:/Users/cnxk/Desktop/dashboard_prac/clean_global_semi_jul.xlsx', engine='xlsxwriter')

semi['Country'].to_excel(writer , sheet_name = 'RAW_Country')
semi['Referring Domain'].to_excel(writer , sheet_name = 'RAW_Referring')
semi['Marketing Channel'].to_excel(writer , sheet_name = 'RAW_Channel')
semi['URL'].to_excel(writer , sheet_name = 'RAW_URL')
semi['Article'].to_excel(writer , sheet_name = 'RAW_Article')
semi['Download'].to_excel(writer , sheet_name = 'RAW_Download')
writer.save()


######################어도비 데이터 그대로 불러오기####################


sheet=['Country','Referring Domain','Marketing Channel','URL','Article','Download']

semi=Series()
for i in range(0,len(sheet)):
    semi[sheet[i]]= pd.read_excel('C:/Users/cnxk/Desktop/dashboard_prac/Semiconductor_China_rawdata_jul.xlsx',

                              sheet_name = sheet[i], 

                              header = 1 )   
    
##dimension열 같은 필요없는 컬럼삭제
    col_not_dimension = []
    for j in semi[sheet[i]].columns:
        if bool(re.match('^Dimension',j)) == False:
            col_not_dimension.append(j)
    semi[sheet[i]]=semi[sheet[i]][col_not_dimension]
    
##자이제 Item컬럼의 빈행, 토탈행, low-traffic 행을 지워야한다.    
    semi[sheet[i]]=semi[sheet[i]].dropna(axis=1,how='all')#첫째열은 전체 빈열->삭제해주기
    semi[sheet[i]]=semi[sheet[i]].dropna(how='any')#NA값들어간 행 지우기
    semi[sheet[i]]=semi[sheet[i]].loc[ (semi[sheet[i]]['Item.1']!='Total'),]
    semi[sheet[i]]=semi[sheet[i]].head(20)

#'Referring Domain'->item.1=total

semi['Country']

writer = pd.ExcelWriter('C:/Users/cnxk/Desktop/dashboard_prac/clean_china_semi_jul.xlsx', engine='xlsxwriter')

semi['Country'].to_excel(writer , sheet_name = 'RAW_Country')
semi['Referring Domain'].to_excel(writer , sheet_name = 'RAW_Referring')
semi['Marketing Channel'].to_excel(writer , sheet_name = 'RAW_Channel')
semi['URL'].to_excel(writer , sheet_name = 'RAW_URL')
semi['Article'].to_excel(writer , sheet_name = 'RAW_Article')
semi['Download'].to_excel(writer , sheet_name = 'RAW_Download')
writer.save()



```
